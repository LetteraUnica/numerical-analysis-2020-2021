```python
import numpy as np
import pylab as pl
from scipy import linalg
```

```python
n = 6

m = 1025

x = np.linspace(0,1,n)
x_plot = np.linspace(0,1,m)
basis = np.array([x_plot**i for i in range(n)]).T
pl.plot(x_plot, basis)
```

```python
from scipy import linalg

def u(x):
    """Runge function"""
    
    return 1/(1+5*(x-0.5)**2)

# Solve linear system with Vandelmond Matrix
uvec = u(x)
van = np.array([x**i for i in range(n)]).T
p_vec = linalg.solve(van, uvec)
```

```python
fun = np.zeros(m)
for i in range(n):
    res = np.ones(m)
    for j in range(n):
        if(j!=i):
            res *= (x_plot-x[j])/(x[i]-x[j])
    fun += res*u(x[i])

pl.plot(x_plot, fun)
pl.plot(x, uvec, "o")
pl.plot(x_plot, u(x_plot))
```

```python
pl.plot(x_plot, basis.dot(p_vec))
pl.plot(x, uvec, "o")
pl.plot(x_plot, u(x_plot))
```

```python
# Study the condition number of Vandermond 

for n in range(1, 25):
    x = np.linspace(0, 1, n)
    van = np.array([x**j for j in range(n)]).T
    print(1/np.min(np.linalg.eig(van)[0]))
```
