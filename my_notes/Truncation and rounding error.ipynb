```python
import numpy as np
import pylab as pl
```

```python
"""Derivative problem"""

def fd(f, x, h):
    return (f(x+h)-f(x))/h

def cfd(f, x, h):
    return (f(x+h)-f(x-h))/(2*h)

def f(x):
    return np.sin(x)

def fprime(x):
    return np.cos(x)

h = 1/2**np.arange(60)
x0 = 1
pl.loglog(h, np.abs(fd(f, x0, h) - fprime(x0)))
```

```python
"""Subtraction problem"""

"""Well conditioned subtraction"""
def f(x, n=50):
    res = np.empty(n)
    res[0] = x
    
    for i in range(1,50):
        x -= x-x/10
        res[i] = x
    
    return res

"""Ill conditioned subtraction"""
def g(x, n=50):
    res = np.empty(n)
    res[0] = x
    
    for i in range(1,50):
        res[i] = x - (x-x/10**i)
    
    return res


pl.plot(np.abs(g(1e100) - f(1e100))/f(1e100))
pl.yscale("log")
```
