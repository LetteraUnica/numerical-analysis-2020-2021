```python
import numpy as np
from scipy.integrate import quad
from numpy.polynomial.legendre import leggauss as leggauss
import pylab as pl
```

```python
def f(x):
    return 1/(1+500*x**2)

N = 100
I = np.empty(N)
for n in range (N):
    q, w = leggauss(n+1)
    I[n] = np.sum(f(q)*w)

x = np.linspace(-1,1,1000)
pl.plot(np.abs(I-quad(f,-1,1)[0]))
pl.yscale("log")
quad(f,-1,1)
```
